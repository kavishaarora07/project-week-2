# cross validation 

```python
def regressor(model,x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=105)
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    
    rmse= mean_squared_error(y_test,pred,squared=False)
    print("mse",rmse)
    
    r2 = r2_score(y_test,pred,multioutput='variance_weighted')
    print("r2 score",r2)
    
    cv_score=(cross_val_score(model,x,y,cv=5).mean())
    print("cv score",cv_score)
    
    result = r2 - cv_score
    print("difference", result)
```

```python
model=LinearRegression()
regressor(model,x,y)
```

```python
model=Lasso(alpha=0.001)
regressor(model,x,y)
```

```python
def classifier(model,x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=105)
    model.fit(x_train,y_train)
    pred=model.predict(x_test)
    
    acc_score= (accuracy_score(y_test,pred))*100
    print("accuracy score",acc_score)
    
    report = classification_report(y_test,pred)
    print("classfication report",report)
    
    cv_score=(cross_val_score(model,x,y,cv=5).mean())*100
    print("cv score",cv_score)
    
    result = acc_score - cv_score
    print("difference", result)
```
