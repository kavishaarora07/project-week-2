```python
df1=df
```

```python
from sklearn.preprocessing import power_transform
df1[cont]=power_transform(df[cont],method='yeo-johnson')
df1=pd.DataFrame(df1,columns=df.columns)
```

```python
s= (df.dtypes == 'object')
cols = list(s[s].index)

print("categorical column")
print(cols)
```

```python
cols.remove('Education')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
for col in cols:
    df1[col]=le.fit_transform(df1[col])
```

```python
df1=df1.drop('Education',axis=1)
```

```python
#dropping education as education_num is already present
```

```python
x=df1.drop('Income',axis=1)
y=df1['Income']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
```

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=448)
```

```python
lr = LogisticRegression()
gb= GaussianNB()
dtc= DecisionTreeClassifier()
svc=SVC()
rf=RandomForestClassifier()
et=ExtraTreesClassifier()
ada=AdaBoostClassifier()
gbc=GradientBoostingClassifier()
models=[lr,gb,dtc,svc,rf,et,ada,gbc]
for i in models:
    print(i)
    i.fit(x_train,y_train)
    pred=i.predict(x_test)
    print(accuracy_score(y_test,pred))
    print(confusion_matrix(y_test,pred))
    print(classification_report(y_test,pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
for i in models:
    print('accuracy score of',i,'is',accuracy_score(y_test,pred))
    cvs=cross_val_score(i,x,y,cv=5)
    print('cross validation score for',i,'model is:',cvs.mean())
    print(' ')
```

```python
#gradient boost is the best
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={"learning_rate":[0.01,0.05,0.1,0.2,0.3,0.5],
       "n_estimators":[5,50,100,200,300,400],
        "max_depth":[3,4,5,6,8]
       }
```

```python
GCV=GridSearchCV(gbc,params,cv=5,scoring='accuracy',n_jobs=-1)
```

```python
GCV.fit(x_train,y_train)
```

```python
GCV.best_estimator_
```

```python
GCV.best_params_
```

```python
pred=GCV.best_estimator_.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
print("accuracy score",round(accuracy_score(y_test,pred)*100,2))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
sns.heatmap(confusion_matrix(y_test,pred),annot=True,fmt='d')
```
